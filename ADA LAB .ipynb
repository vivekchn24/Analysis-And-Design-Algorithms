# 1.3 Implementation and Time analysis of Insertion sort

```python
import time
t_time = time.time()
def insertion_sort(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i-1
        while j >= 0 and key < a[j]:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key


a = []
n = int(input("enter the array size:"))
print("Enter the array elements :")
for i in range(n):
    e = int(input())
    a.append(e)


insertion_sort(a)
print(a)
com_time = time.time()
print("time complexity of insertion sort",com_time - t_time,"seconds")
```

# 2.1 Implementation and Time analysis of Linear search algorithm.

```python
import time
init_time = time.time()
print("\nWelcome to Linear Search Algorithm ")
def linear_search(arr, x):

    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

final_time = time.time()
arr = []
n = int(input("Enter the Require size of Array  :"))
print("Enter the array elements :\n", end="")
for i in range(0, n):
    e = int(input())
    arr.append(e)

x = int(input("Enter the element you want to find :\n"))
index = linear_search(arr, x)
if index == -1:
    print("Element is not found in the array")
else:
    print("Element is found at index -> ", index)
print("Time taken by algorithm to execute is :", final_time - init_time, "seconds")
```

# 2.2 Implementation and Time analysis of Binary search algorithm.

```python
import time
t_time = time.time()

def binary_search(a, key,min,max):
    while min <=max:
        mid = (max + min)
        if a[mid] == key:
            return mid
        elif a[mid] > key:
            max = mid - 1
        elif a[mid] < key:
            min = mid + 1
            return -1

com_time = time.time()
a = []
size = int(input("Enter the size of array you want :\n"))
print("Enter the elements of size ", size, ":")
for i in range(size):
    e = int(input())
    a.append(e)
print(a)
key = int(input("which element you want to fountd in the array?:\n"))

index = binary_search(a,key,0,len(a)-1)
if index == -1:
    print("element doesnt foud")
else:
    print("element found at index numeber:",index)

print("time complexity of the binary search is:" , com_time - t_time , "seconds")
```
